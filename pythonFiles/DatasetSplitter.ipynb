```
from scipy.sparse.linalg import svds
import pandas as pd
import numpy as np
```

```
path = '/content/drive/My Drive/FinalYearProj-2020/DataSets/Movielens/'
csvFilePath='/content/drive/My Drive/FinalYearProj-2020/csvFiles/'
ratings = pd.read_csv(path+'ratings.csv')
ratings.head()
```

```
np.random.seed = 42
msk = np.random.rand(len(ratings)) < 0.8
trn = ratings[msk]
vald = ratings[~msk]
trn.shape
```

```
trn.to_csv(path+"trainset.csv", index=False)
vald.to_csv(path+"testset.csv",index=False)
```

```
UserMovieRatingTableTrainset = trn.pivot_table(index='userId',columns='movieId',values='rating')
# UserMovieRatingTableTrainset.head(10)
UserMovieRatingTableTrainset.to_csv(csvFilePath+"UserMovieRatingsTrainset.csv")
```

```
UserMovieRatingTableTestset = vald.pivot_table(index='userId',columns='movieId',values='rating')
UserMovieRatingTableTestset.head(10)
UserMovieRatingTableTestset.to_csv(csvFilePath+"UserMovieRatingsTestset.csv")
```
