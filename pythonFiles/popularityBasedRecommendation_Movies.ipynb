```python
movieRatings=pd.merge(movies,ratings,on="movieId")
movieRatings
```

```python
movieRatings= movieRatings.dropna()
movieRatings
```

```python
movie_ratingCount = (movieRatings.groupby(by=['movieId'])['rating'].count().reset_index().rename(columns = { 'rating':'totalRatingCount'})[['movieId','totalRatingCount']])
movie_ratingCount
```

```python
dup = movie_ratingCount.reset_index().drop_duplicates(subset=['movieId'],keep='first').set_index('index')
dup
```

```python
rating_with_totalRatingCount = movieRatings.merge(movie_ratingCount, on = 'movieId').reset_index().drop_duplicates(subset=['movieId'],keep='first').set_index('index').drop(['userId','rating'],axis=1)
rating_with_totalRatingCount
# rating_with_totalRatingCount_Unique = rating_with_totalRatingCount.reset_index().drop_duplicates(subset=['movieId'],keep='first').set_index('index')
# rating_with_totalRatingCount_Unique
```

```python
#giving some popularity thresold to discard less popular books
popularity_thresold = 50
rating_popular_movie = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_thresold')
rating_popular_movie
```

```python
# movieRatings=rating_popular_movie
movieRatings
```

```python
movie_ratingAvg = (movieRatings.groupby(by=['movieId'])['rating'].mean().reset_index().rename(columns = { 'rating':'totalRatingAvg'})[['movieId','totalRatingAvg']])
movie_ratingAvg
```

```python
movie_ratingAvg=pd.merge(rating_popular_movie,movie_ratingAvg,on='movieId')
movie_ratingAvg
```

```python
# zscore = pd.DataFrame([])
# book_ratingAvg=book_ratingAvg.drop(['Z_Score'],axis=1)
movie_ratingAvg['z_score'] = (movie_ratingAvg['totalRatingAvg'] - movie_ratingAvg['totalRatingAvg'].mean())/movie_ratingAvg['totalRatingAvg'].std(ddof=0)
movie_ratingAvg
```

```python
# movie_ratingAvg['score']=(movie_ratingAvg['timestamp']*0.00000001)*(movie_ratingAvg['z_score']*0.99999999)
# movie_ratingAvg
# popularityBasedFilteredTable.sort_values(by="score",ascending=False).head(10)

maxTime = movie_ratingAvg['timestamp'].max()
maxRated = movie_ratingAvg['totalRatingCount'].max()
movie_ratingAvg['score']=(movie_ratingAvg['timestamp']/maxTime)*(movie_ratingAvg['z_score'])
movie_ratingAvg
```

```python
popularityBasedFilteredTable = movie_ratingAvg.sort_values(by="score",ascending=False)
popularityBasedFilteredTable.head(10)
```
