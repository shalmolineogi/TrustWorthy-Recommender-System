```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
Avg_TrustMatrix = pd.read_csv('/content/drive/My Drive/My Fypr/csvFiles/Average_Trust_Matrix.csv',index_col = 0)
Avg_TrustMatrix.index.name = 'userIdRow'
Avg_TrustMatrix.columns.name = 'userIdCol'
Avg_TrustMatrix.head(10)
```

```
keyList = list(Avg_TrustMatrix.keys())
n = len(keyList)
print(n,keyList)
#keyList = list(map(int,keyList))
#print(keyList)
```

```
def makeTrustList(TrustMatrix,trustval):
    trustedList = []
    for i in range(n):
        trusted = []
        lst = TrustMatrix[TrustMatrix[keyList[i]] > trustval].index.tolist()
        for el in lst:
            val = keyList.index(str(el))
            if val >= i : 
                trusted.append(val)
        trustedList.append(trusted)
    return trustedList
```

```
def makegroup(currentGroup):
    grpLen = len(currentGroup)
    userGroup = []
    notDone = [i for i in range(n)]
    for i in range(grpLen):
        groupTemp = []
        user = currentGroup[i][0]
        if user in notDone:
            notDone.remove(user)
            groupTemp.append(user)
            for j in range(1,len(currentGroup[i])):
                ele = currentGroup[i][j]
                if ele in notDone:
                    notDone.remove(ele)
                    groupTemp.append(ele)
                    for k in range(1,len(trustList[ele])):
                        tmpele = trustList[ele][k]
                        if tmpele in notDone:
                            notDone.remove(tmpele)
                            groupTemp.append(tmpele)
        if len(groupTemp) != 0:
            userGroup.append(groupTemp) 
    return userGroup
```

```
trustList = makeTrustList(Avg_TrustMatrix,.999)
#print(trustList)
dividedGroup = makegroup(trustList)
print('Number of groups are : ',len(dividedGroup))
for i in dividedGroup:
    print(i)
```
