```python
import numpy as np
import pandas as pd
```

```python
UserRatings = pd.read_csv('ratings.csv',)
UserRatings.head()
```

```python
Movies = pd.read_csv("movies.csv")
Movies.head()
```

```python
DataSet = pd.merge(UserRatings,Movies,on='movieId')
DataSet.head()
```

```python
RatingCount = DataSet.groupby('title')['rating'].count().reset_index().rename(columns = {'rating':'rating_count'})
RatingCount.head()
```

```python
MeanRatings = DataSet.groupby('title')['rating'].mean().reset_index().rename(columns = {'rating':'mean_rating'})
MeanRatings.head()
```

```python
DataSet = pd.merge(DataSet,RatingCount,on='title')
DataSet = pd.merge(DataSet,MeanRatings,on='title')
DataSet.head()
```

```python
UserMovieRatingTable = DataSet.pivot_table(index='userId',columns='title',values='rating')
UserMovieRatingTable.head()
```

```python
MovieUserRatingTable = UserMovieRatingTable.transpose().fillna(0)
MovieUserRatingTable.head()
```

```python
OneMovieRating = UserMovieRatingTable['Toy Story (1995)']
OneMovieRating.head()
```

```python
SimilarMovies = UserMovieRatingTable.corrwith(OneMovieRating)
CorrelationStatus = pd.DataFrame(SimilarMovies,columns = ['Correlation'])
```

```python
CorrelationStatus.sort_values('Correlation',ascending=False).head()
```

```python
MeanRatings = pd.DataFrame(DataSet.groupby('title')['rating'].mean())
MeanRatings['rating_count'] = pd.DataFrame(DataSet.groupby('title')['rating'].count())
MeanRatings.head()
```

```python
CorrelationStatus = CorrelationStatus.join(MeanRatings['rating_count'])
```

```python
CorrelationStatus.head(10)
```

```python
CorrelationStatus[CorrelationStatus['rating_count']>100].sort_values('Correlation',ascending=False).head()
```
