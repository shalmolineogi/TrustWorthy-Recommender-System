```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
UserBookRating = pd.read_csv('/content/drive/My Drive/FinalYearProj-2020/csvFiles/UserBookRating.csv',index_col=0)
UserBookRating.columns.name = 'BookTitle'
UserBookRating.head()
```

```python
UserBookRating.describe()
```

```python
zscore = pd.DataFrame([])
m,n = UserBookRating.shape
print(m,n)
for i in range(n):
    zscore[UserBookRating.columns[i]] = (UserBookRating.iloc[:,i] - UserBookRating.iloc[:,i].mean())/UserBookRating.iloc[:,i].std(ddof=0)
zscore.head(10)
```

```python
zscore.describe()
```
